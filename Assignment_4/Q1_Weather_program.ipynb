```python
import requests

def get_coordinates(city, state, country, api_key):
    base_url = "http://api.openweathermap.org/geo/1.0/direct"
    params = {
        'q': f"{city},{state},{country}",
        'limit': 1,
        'appid': api_key
    }
    
    try:
        response = requests.get(base_url, params=params)
        data = response.json()
        
        if data:
            lat = data[0]['lat']
            lon = data[0]['lon']
            return lat, lon
        else:
            return None, None
    except Exception as e:
        print("Error:", str(e))
        return None, None

def get_weather(lat, lon, api_key):
    base_url = "http://api.openweathermap.org/data/2.5/weather"
    params = {
        'lat': lat,
        'lon': lon,
        'appid': api_key,
        'units': 'metric' 
    }
    
    try:
        response = requests.get(base_url, params=params)
        data = response.json()
        
        if data['cod'] == 200:
            weather_info = {
                'Temperature': data['main']['temp'],
                'Feels Like': data['main']['feels_like'],
                'Description': data['weather'][0]['description']
            }
            return weather_info
        else:
            return {'error': data['message']}
    except Exception as e:
        return {'error': str(e)}

def main():
    api_key = '975e432c37db45d3858815fd7c043f7b'  
    city = input("Enter city name(like: Boston): ")
    state = input("Enter state code(like: MA): ")
    country = input("Enter country code(like: US): ")
    
    lat, lon = get_coordinates(city, state, country, api_key)
    if lat is not None and lon is not None:
        weather_info = get_weather(lat, lon, api_key)
        if 'error' in weather_info:
            print("Error:", weather_info['error'])
        else:
            print("Weather Information:")
            for key, value in weather_info.items():
                print(f"{key}: {value}")
    else:
        print("Unable to retrieve coordinates.")

if __name__ == "__main__":
    main()
```
